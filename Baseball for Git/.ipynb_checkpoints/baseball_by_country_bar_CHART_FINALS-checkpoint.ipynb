```python
%matplotlib inline
import pandas as pd
import numpy as np
#for graph
import matplotlib.pyplot as plt



baseball = pd.read_csv("master.csv")
baseball1 = pd.read_csv("appearances_yearid_playerid.csv")

#copy the sets to not accidentally change them
A = baseball[:]
#print(A.head(5))
B = baseball1.copy()
#print(B.head(5))

#B columbs to the left and add the A columns to the right side..
E = pd.merge(A, B, on='playerID')
print(E.info())
#see what the info looks like to see if i have alot of nans
#i can see that some of the birthcountries must be missing since there are fewer birthcoutnries than players
#so about 87 players seem to lack countries...
#there are additional player IDs because it shows a playersID for each year they played, I assume

```

```python

print(E.tail(10))


```

```python
#for the purpose of this Im going to slim down to eliminate debut and final game...
F = E[["playerID", "birthCountry", "yearID", "teamID", "lgID"]]
F2 = F[["birthCountry","yearID","playerID"]]
F.tail(10)
```

```python

#drop rows that contain null values
F19 = F2.dropna()
F20 = F19.loc[F2["birthCountry"] !="USA"]
F21 = F20.replace("CAN","Canada")
F22 = F21.replace("P.R.", "Puerto Rico")
F23 = F22.replace("D.R.", "Dominican Republic")
#big papi's birthday is in 1975
F24 = F23.loc[F23["yearID"]> 1974]

#making top five for line chart
#need to say Im leaving in Puerto Rico just for fun..
#to filter out only the top countries
countrylist = ['Canada','Puerto Rico',"Venezuela","Mexico","Cuba","Dominican Republic"]
F25 = F24[F24['birthCountry'].isin(countrylist)]
#exporting to use as a csv for publication


F25.columns = [["birthCountry","Year","playerID"]]

#these are the top non uSA countries
##D.R.              3726
#P.R.              1854
#Venezuela         1835
#Canada            1103
#Cuba              1077
#Mexico             714
#Panama             367
#Japan              265
#Ireland            242
#United Kingdom     231
#Germany            192
```

```python
#this gets the bigger group, the countries with the most players SINCE 1975, not all time, 
#they are different
#separate out those top values
#F24.groupby("birthCountry").size().sort_values().tail(10)\
#.sort_values(ascending=False).plot(kind='bar', title ="Top Birthplaces For Players Since 1975",\
                                  # fontsize=12, figsize=(8, 4))

F291 = F24.groupby("birthCountry").size().sort_values().tail(10)\
.sort_values(ascending=False)
#how to use this 

fig, ax = plt.subplots(figsize=(8,10))


#,title ="Top Birthplaces For Players Since 1975",fontsize=20
ax = F291.plot(kind='bar')
ax.set_ylabel('Number of Players',fontsize=20)
ax.set_xlabel('')   
ax.spines["top"].set_visible(False)    
#ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
#ax.spines["left"].set_visible(False) 
#plt.grid(False)
plt.title("Players Born Outside the U.S. Since 1975",fontsize=20)
plt.xticks(fontsize=18)  
plt.yticks(fontsize=18)
fig.tight_layout()

plt.show()

plt.savefig("baseball_barChart.svg")

#(kind="bar", figsize=(12,8))
#(kind='bar', title ="Top Birthplaces For Players Since 1975", fontsize=12, figsize=(8, 4))



#plot them in name order rather than size order
#bar1 = F24

#ax = df[['V1','V2']].plot(kind='bar', title ="V comp", figsize=(15, 10), legend=True, fontsize=12)
#barChart = plt.plot(bar1)
#plt.ylabel("Number of Players",fontsize=18)
#bar1.plot(kind="bar", figsize=(12,8), ylabel)


```

```python

yearCountry = F25.groupby(["Year","birthCountry"])
yearCountrySize = yearCountry.size().unstack()
#yearCountrySize.to_csv('baseballLineChartUSE.csv')
#getting rid of plot frame lines, from http://www.randalolson.com/2014/06/28/how-to-make-beautiful-data-visualizations-in-python-with-matplotlib/
   
countryBySize = yearCountry["birthCountry"].size()

# i dont know how to set colors
#plt.gca().set_prop_cycle(['rgb(166,206,227)','rgb(31,120,180)','rgb(178,223,138)'])
#def colors():
  #  coli = ["b","g","y","k","c"]
   # for index in range(len(coli)):
   # return coli[index]

#plot, lw is line width, made thicker for thumbnail
plots = plt.plot(yearCountrySize, lw=2.5)

#plt.show()
#make sure list of country is in order of playerid size..
#cleverly it figures out the legend without much help..
plt.legend(plots,(yearCountrySize),loc="upper left",
                  framealpha=0.25,
                  #prop={"size":"small","family":"monospace"}
                  fontsize="x-large")

#set size of chart
plt.gcf().set_size_inches(12,10)
plt.title("Where Do Baseball Players Come From?",fontsize=20)
#plt.xlabel("Year")
plt.ylabel("Number of Players",fontsize=18)
#can turn the grid on or off - http://www.randalolson.com/2014/06/28/how-to-make-beautiful-data-visualizations-in-python-with-matplotlib/
#setting visibility of box around graph
ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False) 
plt.grid(False)
#setting the length and width to the min and max of the player number and year data
plt.ylim(0, 165)   
plt.xlim(1975, 2015)  

#plot tick size
plt.xticks(fontsize=16)  
plt.yticks(fontsize=16)
#adds light gray ticks across using max and min of players and years, 10 apart
for y in range(10, 165, 10):    
    plt.plot(range(1975, 2015), [y] * len(range(1975, 2015)), "--", lw=0.5, color="black", alpha=0.3)    
#adds copyright
plt.figtext(0.99,-0.05, u'\u00a9 HLW', 
            ha="right", va="bottom")

#adds sourcing for data
plt.figtext(0.99,-0.07, "Data source: Sean Lahman baseball database",
           ha="right", va="bottom", fontsize = 8)

#keeps layout tight!
plt.tight_layout()
plt.savefig("lineChart_thicker_use.svg")

#sns.palplot(sns.color_palette("colorblind", 10))
#fig size can stretch or shrink it, first no is length, second is width(height)
#yearCountrySize.plot(kind="line", figsize=(16,8))
#yea! after a million years, did it! in a manner of speaking..
```
