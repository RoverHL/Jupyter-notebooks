in 2017, unlike other years, they had some extra ( in the csv, so had to add .replace( '[\(]','', regex=True ) i addition to earlier versionsthat removing the $ -- not sure this is the best way but it works :)

```python
d_2017['REGULAR'] = (d_2017['REGULAR'].replace( '[\$,)]','', regex=True ).replace( '[\(]','', regex=True ).astype(float)).round()
d_2017['RETRO'] = (d_2017['RETRO'].replace( '[\$,)]','', regex=True ).replace( '[\(]','', regex=True ).astype(float)).round()
d_2017['OTHER'] = (d_2017['OTHER'].replace( '[\$,)]','', regex=True ).replace( '[\(]','', regex=True ).astype(float)).round()
d_2017['OVERTIME'] = (d_2017['OVERTIME'].replace( '[\$,)]','', regex=True ).replace( '[\(]','', regex=True ).astype(float)).round()
d_2017['INJURED'] = (d_2017['INJURED'].replace( '[\$,)]','', regex=True ).replace( '[\(]','', regex=True ).astype(float)).round()
d_2017['DETAIL'] = (d_2017['DETAIL'].replace( '[\$,)]','', regex=True ).replace( '[\(]','', regex=True ).astype(float)).round()
d_2017['QUINN/EDUCATION INCENTIVE'] = (d_2017['QUINN/EDUCATION INCENTIVE'].replace( '[\$,)]','', regex=True ).replace( '[\(]','', regex=True ).astype(float)).round()
d_2017['TOTAL EARNINGS'] = (d_2017['TOTAL EARNINGS'].replace( '[\$,)]','', regex=True ).replace( '[\(]','', regex=True ).astype(float)).round()
```
